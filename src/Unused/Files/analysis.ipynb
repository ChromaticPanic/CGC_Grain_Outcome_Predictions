```python
for df in dfList:
    df["rolavgMean"] = df["MeanTemp"].rolling(window=5).mean()
    df["rolavgMin"] = df["MinTemp"].rolling(window=5).mean()
    df["rolavgMax"] = df["MaxTemp"].rolling(window=5).mean()
```

```python
fig, axes = plt.subplots(3, 3, figsize=(30, 20))

j = 0
k = 0
for df in dfList:
    sns.scatterplot(
        data=df, x="YMD", y="rolavgMean", ax=axes[j, k], color="green", size=0.2
    )
    sns.scatterplot(
        data=df, x="YMD", y="rolavgMin", ax=axes[j, k], color="blue", size=0.2
    )
    sns.scatterplot(
        data=df, x="YMD", y="rolavgMax", ax=axes[j, k], color="orange", size=0.2
    )
    k += 1
    if k == 3:
        j += 1
        k = 0


plt.show()
```
